##LINEAR REGRESSION AND GRADIENT DESCENT

### Mean Squarred Error (MSE)

```python
def MSE(Y,Y_PRED):
  n = len(Y)
  sum = 0
  for k in range(0, len(Y)):
    sum += (Y[k]-Y_PRED[k])**2
  return sum/n
```

```python
def partialDerivMSE_B1(X,errors):
  n= len(X)
  sum=0
  for i in range(0,len(X)):
    sum += -2*X[i][0] * errors[i]
  return sum/n
```

```python
def partialDerivMSE_B0(errors):
  n=len(errors)
  sum = 0
  for i in range(0,len(errors)):
    sum += -2*errors[i]
  return sum/n
```

### Gradient descent algorithm

```python
def gradientDescent(X_train, X_test, y_train, y_test,epochs, alpha, beta_0,beta_1):
  
  betas0 = []
  betas1 = []
  MSE_train = []
  MSE_test = []

  for i in range(0,epochs):

    #Update Parameters
    errors_train = []
    for xti in range(0,len(X_train)):
      xi = X_train[xti][0]
      y_true = y_train[xti]
      y_pred = beta_0 + beta_1*xi
      error = y_true - y_pred
      errors_train.append(error)

    dB0 = partialDerivMSE_B0(errors_train)
    dB1 = partialDerivMSE_B1(X_train,errors_train)
    beta_0 -= alpha * dB0
    beta_1 -= alpha * dB1

    betas0.append(beta_0)
    betas1.append(beta_1)

    #Compute MSE on Train and Test
    pred=[]
    for xti in range(0,len(X_train)):
      xi = X_train[xti][0]
      y_pred = beta_0 + beta_1*xi
      pred.append(y_pred)
    
    lossTrain = MSE(y_train,pred)
    MSE_train.append(lossTrain)
    
    pred=[]
    for xti in range(0,len(X_test)):
      xi = X_test[xti][0]
      y_pred = beta_0 + beta_1*xi
      pred.append(y_pred)
    lossTest = MSE(y_test,pred)
    MSE_test.append(lossTest)

    if (i % 10000 == 0 or i==0 or i==epochs-1):
      print("Epoch "+str(i)+'/'+str(epochs)+" [==============================] "+"Loss: "+str(lossTrain) + " - Val_loss: "+str(lossTest)+" B0: ",str(beta_0)," B1: ",str(beta_1))
  
  return [MSE_train, MSE_test,betas0,betas1]
```

#### Display the optimisation

```python
def displayHistory(epoch_min,epoch_max,history):
  

  epochs = np.arange(epoch_min,epoch_max,step=1)
  sMSE_Train = history[0][epoch_min:epoch_max]
  sMSE_Val = history[1][epoch_min:epoch_max]
  sBetas0 = history[2][epoch_min:epoch_max]
  sBetas1 = history[3][epoch_min:epoch_max]

  #MSE Test / Train
  fig, (ax1) = plt.subplots(1,figsize=(15, 4))
  ax1.plot(epochs,sMSE_Train,color="gray",label=("MSE Train"))
  ax1.plot(epochs,sMSE_Val,color="purple",label=("MSE Test"))
  ax1.set_title('MSE')
  ax1.legend()
  ax1.grid()
  

  #Beta 0 and Beta 1
  fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(15, 4))
  ax1.plot(epochs, sBetas0,color="gray")
  ax1.set_title('Beta 0')
  ax1.grid()
  ax2.plot(epochs, sBetas1,color="gray")
  ax2.set_title('Beta 1')
  ax2.grid()

  
  plt.show()  
```
