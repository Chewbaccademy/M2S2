```python
import numpy
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.utils import np_utils
numpy.random.seed(7)
```

# PREPARATION DES DONNEES

```python
alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

#Caractere vers entiers
caractere_vers_entier = dict((c, i) for i, c in enumerate(alphabet))

#Entier vers caracteres
entier_vers_caractere = dict((i, c) for i, c in enumerate(alphabet))
```

```python
caractere_vers_entier
```

```python
entier_vers_caractere
```

### DONNEES A PREDIRE

```python
taille_sequence = 1 #Contexte d'apprentissage d'une lettre
dataX = []
dataY = []
for i in range(0, len(alphabet) - taille_sequence, 1):
	sequence_entree = alphabet[i:i + taille_sequence]
	sequence_sortie = alphabet[i + taille_sequence]
	dataX.append([caractere_vers_entier[caractere] for caractere in sequence_entree])
	dataY.append(caractere_vers_entier[sequence_sortie])
	print("Si :",sequence_entree, ' Prédire :', sequence_sortie)
```

```python
dataX
```

```python
dataY
```

```python
#Format necessaire pour les LSTM
X = numpy.reshape(dataX, (len(dataX), taille_sequence, 1))
X.shape
```

```python
#Normalisation
X = X / float(len(alphabet))
X
```

```python
#One Hot Encoding
y = np_utils.to_categorical(dataY)
y
```

# LSTM 1

### APPRENTISSAGE

```python
X.shape[1]
```

```python
 X.shape[2]
```

```python
model = Sequential()

#32 Unités
model.add(LSTM(32, input_shape=(X.shape[1], X.shape[2])))

#Sortie de 25 neurones (correspondant au 25 lettres)
model.add(Dense(y.shape[1], activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, y, epochs=500, batch_size=1, verbose=2)
```

### EVALUATION / PREDICTIONS

```python
scores = model.evaluate(X, y, verbose=0)
print("Justesse (accuracy): %.2f%%" % (scores[1]*100))
```

```python
for lettre in dataX:

  #On formate la lettre pour que celle-ci soit utilisable dans le LSTM
  x = numpy.reshape(lettre, (1, len(lettre), 1))
 
  #Normalisation
  x = x / float(len(alphabet))
 
  #Prediction
  prediction = model.predict(x, verbose=0)
	
  #Classe maximale
  index = numpy.argmax(prediction)
	
  sequence_entree = [entier_vers_caractere[valeur] for valeur in lettre]
  sequence_sortie = entier_vers_caractere[index]
  
  print(sequence_entree, "->", sequence_sortie)
```

Nous pouvons voir que ce problème est  difficile pour le réseau.

La raison en est que les unités LSTM n'ont aucun contexte avec lequel travailler. (Lettre par lettre : Pas de contexte)

Chaque modèle d'entrée-sortie est présenté au réseau dans un ordre aléatoire et l'état du réseau est réinitialisé après chaque modèle (chaque lot où chaque lot contient un modèle).

Il s'agit d'un abus de l'architecture du réseau LSTM, qui le traite comme un perceptron multicouche standard.

Ensuite, essayons de formuler le problème différemment afin de fournir au réseau plus de séquences à partir desquelles il peut apprendre.


# LSTM 2 : UN PEU DE CONTEXTE

```python
alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

#Caractere vers entiers
caractere_vers_entier = dict((c, i) for i, c in enumerate(alphabet))

#Entier vers caracteres
entier_vers_caractere = dict((i, c) for i, c in enumerate(alphabet))
```

```python
taille_sequence = 3 #Contexte de 3 lettres
dataX = []
dataY = []
for i in range(0, len(alphabet) - taille_sequence, 1):
	sequence_entree = alphabet[i:i + taille_sequence]
	sequence_sortie = alphabet[i + taille_sequence]
	dataX.append([caractere_vers_entier[caractere] for caractere in sequence_entree])
	dataY.append(caractere_vers_entier[sequence_sortie])
	print("Si :",sequence_entree, ' Prédire :', sequence_sortie)
```

```python
#PARAMETRES :
#Datas
#Samples : Nb Observations : 23
#Time Steps = Qté de pas de temps : 1
#Features : Quantité d'éléments à chaque time step

X = numpy.reshape(dataX, (len(dataX), 1, taille_sequence))
X

#Exemple ABC
# len(dataX) : 23 Nombres d'observation
# 1 : Time steps de 1
# taille_sequence : 3

#On obtient donc un tableau de 23 lignes

#> SEQUENCE : 1 TIME STEP DE MULTIPLES FEATURES
# "ABC"

```

```python
#ONE HOT CODING
y = np_utils.to_categorical(dataY)
```

```python

#Normalisation
X = X / float(len(alphabet))
```

```python
model = Sequential()

#32 Unités
model.add(LSTM(32, input_shape=(X.shape[1], X.shape[2])))

#Sortie de 25 neurones (correspondant au 25 lettres)
model.add(Dense(y.shape[1], activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, y, epochs=500, batch_size=1, verbose=2)
```

```python
scores = model.evaluate(X, y, verbose=0)
print("Justesse (accuracy): %.2f%%" % (scores[1]*100))
```

```python
for lettre in dataX:

  #On formate la lettre pour que celle-ci soit utilisable dans le LSTM
  x = numpy.reshape(lettre,(1,1,len(lettre)))
 
  #Normalisation
  x = x / float(len(alphabet))
 
  #Prediction
  prediction = model.predict(x, verbose=0)
	
  #Classe maximale
  index = numpy.argmax(prediction)
	
  sequence_entree = [entier_vers_caractere[valeur] for valeur in lettre]
  sequence_sortie = entier_vers_caractere[index]
  
  print(sequence_entree, "->", sequence_sortie)
```

# PREDIRE DES "PHRASES"

Pour le moment notre architecture est capable de prédire une lettre à partir d'une séquence de 3 lettres.

Mais si on lui présente une seule lettre ou deux lettres, il ne peut pas la prédire.

```python
alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

#Caractere vers entiers
caractere_vers_entier = dict((c, i) for i, c in enumerate(alphabet))

#Entier vers caracteres
entier_vers_caractere = dict((i, c) for i, c in enumerate(alphabet))
```

```python
nbPhases = 1000
nbCaracteresMaxParPhrase = 5
```

```python
dataX = []
dataY = []
for i in range(nbPhases):

  #Choix d'un numéro de caractere au hasard pour débuter la phrase
  caractereDebutPhrase = numpy.random.randint(len(alphabet)-2)
 
  #Choix d'un numéro de caractere au hasard pour finir la phrase
  caractereFinPhrase = numpy.random.randint(caractereDebutPhrase, min(caractereDebutPhrase+nbCaracteresMaxParPhrase,len(alphabet)-1))
 

  sequence_entree = alphabet[caractereDebutPhrase:caractereFinPhrase+1]
  sequence_sortie = alphabet[caractereFinPhrase + 1]
	
  dataX.append([caractere_vers_entier[caractere] for caractere in sequence_entree])
  dataY.append(caractere_vers_entier[sequence_sortie])
  print("Si :",sequence_entree, ' Prédire :', sequence_sortie)
 
```

```python
from keras.preprocessing.sequence import pad_sequences

#On fait un padding pour les phrases inférieures à 5 caracteres
X = pad_sequences(dataX, maxlen=nbCaracteresMaxParPhrase, dtype='float32')
X
```

```python
#Reshape, normalisation, on hot encoding
X = numpy.reshape(X, (X.shape[0], nbCaracteresMaxParPhrase, 1))
X = X / float(len(alphabet))
y = np_utils.to_categorical(dataY)
```

```python
X[0] #Nous diposons bien d'une séquence de 5 caracteres
```

```python
y[0]
```

```python
model = Sequential()
model.add(LSTM(32, input_shape=(X.shape[1], 1)))
model.add(Dense(y.shape[1], activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, y, epochs=500, verbose=1)

```

```python
for i in range(20):
	
  #Creation de la phrase dans le dataset
  choixNumSequenceDansDataset = numpy.random.randint(len(dataX))
  
  #Utilisation de la sequence
  sequenceSelectionnee = dataX[choixNumSequenceDansDataset]
  print(sequenceSelectionnee)
  x = pad_sequences([sequenceSelectionnee], maxlen=nbCaracteresMaxParPhrase, dtype='float32')
  x = numpy.reshape(x, (1, nbCaracteresMaxParPhrase, 1))
  x = x / float(len(alphabet))
	
  #Prediction
  prediction = model.predict(x, verbose=0)
  index = numpy.argmax(prediction)
  

  sequence_sortie = entier_vers_caractere[index]
  sequence_entree = [entier_vers_caractere[lettre] for lettre in sequenceSelectionnee]
  print(sequence_entree, "->", sequence_sortie)
```
