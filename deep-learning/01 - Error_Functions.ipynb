```python
import numpy as np
import matplotlib.pyplot as plt
```

## CONVEXITY

$$\forall x,y \in I^2, \forall t \in [0,1], f(\theta x+(1-\theta)y) \leq \theta f(x) + (1-\theta)f(y)$$

The objective of learning is to minimize an error function.
How to understand the examples below?
- Let's imagine an algorithm 
- Several iterations are performed
- At each iteration, we compute the error between the y-value and the prediction using an error function
- At each iteration, we suppose that the algorithm adjusts its parameters in order to minimize its error

##MAE

$$MAE = \frac{\sum^n_{i=1}\vert y_i - \hat{y_i} \vert}{N}$$

```python
def mae(y_true, y_pred):
  N= len(y_true)
  sum = 0
  for i in range (0,len(y_true)):
    sum += abs(y_true[i]-y_pred[i])
  return sum/N
```

```python
#Expected value
y_true = 1

#Predictions at each iteration
#-8;-7.99;-7.98...
I = [-8,8]
predictions = np.arange(I[0],I[1],0.01)

#Errors estimation with the error function
maeResults = []
for y_pred in predictions:
  maeResults.append(mae([y_true],[y_pred]))
```

```python
plt.figure(figsize=(6, 6))
plt.plot(predictions,maeResults,color="black",linewidth=0.85)
plt.grid()
plt.xlabel('y_hat', fontsize=14)
plt.ylabel('MAE', fontsize=14)
plt.show()
```

```python
##Left side and right side of the inequality
##-> Don't forget to round the result :
##0.06000000005 != 0.06000000000
def leftSide (y_true,x,y,t):
  a = (1-t)*y+t*x
  return round(mae([y_true],[a]),2)

def rightSide (y_true,x,y,t):
  a = (t*mae([y_true],[x]))+((1-t)*mae([y_true],[y]))
  return round(a,2)


#Convexity
T  = np.arange(0,1,0.02)
X = np.arange(I[0],I[1],0.01)
Y = X[::-1]

nbTests = 0
nbPositivesTests = 0
for i in range(0,len(X)):
  x = X[i]
  y = Y[i]
  for t in T:
    nbTests+=1
    left = leftSide(y_true,x,y,t)
    right = rightSide(y_true,x,y,t)
    if (left < right or left==right):
      nbPositivesTests +=1
if (nbTests == nbPositivesTests):
  print("-> The MAE function is convex")
else:
  print("-> The MAE function is not convex")
```

## MSE

$$MSE = \frac{\sum^n_{i=1}(y_i - \hat{y_i})^2}{N}$$

```python
def mse(y_true, y_pred):
  N= len(y_true)
  sum = 0
  for i in range (0,len(y_true)):
    sum += (y_pred[i]-y_true[i])**2
  return sum/N
```

```python
y_true = 3
I = [-8,8]
predictions = np.arange(I[0],I[1],0.01)

mseResults = []
for y_pred in predictions:
  mseResults.append(mse([y_true],[y_pred]))

plt.figure(figsize=(6, 6))
plt.plot(predictions,mseResults,color="black",linewidth=0.85)
plt.grid()
plt.xlabel('y_hat', fontsize=14)
plt.ylabel('MSE', fontsize=14)
plt.show()
```

```python
##Left side and right side of the inequality
##-> Don't forget to round the result :
##0.06000000005 != 0.06000000000
def leftSide (y_true,x,y,t):
  a = (1-t)*y+t*x
  return round(mse([y_true],[a]),2)

def rightSide (y_true,x,y,t):
  a = (t*mse([y_true],[x]))+((1-t)*mse([y_true],[y]))
  return round(a,2)


#Convexity
T  = np.arange(0,1,0.02)
X = np.arange(I[0],I[1],0.01)
Y = X[::-1]

nbTests = 0
nbPositivesTests = 0
for i in range(0,len(X)):
  x = X[i]
  y = Y[i]
  for t in T:
    nbTests+=1
    left = leftSide(y_true,x,y,t)
    right = rightSide(y_true,x,y,t)
    if (left < right or left==right):
      nbPositivesTests +=1
if (nbTests == nbPositivesTests):
  print("-> The MSE function is convex")
else:
  print("-> The MSE function is not convex")
```

##RMSE

$$RMSE = \sqrt\frac{\sum^n_{i=1}(y_i - \hat{y_i})^2}{N}$$

```python
def rmse(y_true,y_pred):
  N= len(y_true)
  sum = 0
  for i in range (0,len(y_true)):
    sum += (y_pred[i]-y_true[i])**2
  return np.sqrt(sum/N)
```

```python
y_true = 3
I = [-8,8]
predictions = np.arange(I[0],I[1],0.01)

rmseResults = []
for y_pred in predictions:
  rmseResults.append(rmse([y_true],[y_pred]))

plt.figure(figsize=(6, 6))
plt.plot(predictions,rmseResults,color="black",linewidth=0.85)
plt.grid()
plt.xlabel('y_hat', fontsize=14)
plt.ylabel('RMSE', fontsize=14)
plt.show()
```

```python
##Left side and right side of the inequality
##-> Don't forget to round the result :
##0.06000000005 != 0.06000000000
def leftSide (y_true,x,y,t):
  a = (1-t)*y+t*x
  return round(rmse([y_true],[a]),2)

def rightSide (y_true,x,y,t):
  a = (t*rmse([y_true],[x]))+((1-t)*rmse([y_true],[y]))
  return round(a,2)


#Convexity
T  = np.arange(0,1,0.02)
X = np.arange(I[0],I[1],0.01)
Y = X[::-1]

nbTests = 0
nbPositivesTests = 0
for i in range(0,len(X)):
  x = X[i]
  y = Y[i]
  for t in T:
    nbTests+=1
    left = leftSide(y_true,x,y,t)
    right = rightSide(y_true,x,y,t)
    if (left < right or left==right):
      nbPositivesTests +=1
if (nbTests == nbPositivesTests):
  print("-> The RMSE function is convex")
else:
  print("-> The RMSE function is not convex")
```

###Binary cross entropy

$$-y_i \times ln(\hat{p_i})-(1-y_i)ln(1-\hat{p_i})$$

$$p_i = \text{is the probability of class 1 given in the prediction made by the model}$$
$$1-p_i = \text{is the probability of class 0}$$

The reason behind using the log value is, the log value offers less penalty for small differences between predicted probability and corrected probability

Cross-entropy loss measures the difference between the discovered probability distribution of a machine learning classification model and the predicted distribution.

- Binary cross entropy : 2 classes
- Categorical cross entropy : More than 2 classes

```python
def BinaryCrossEntropy(prob_y_true,prob_y_pred):
  return -prob_y_true*np.log(prob_y_pred)+(1-prob_y_true)*np.log(1-prob_y_pred)
```

```python
probability = np.arange(0.01,1,0.01)
y_expected_proba = 1
b = []
for y_pred_proba in probability:
  b.append(BinaryCrossEntropy(y_expected_proba,y_pred_proba))


plt.figure(figsize=(6, 6))
plt.plot(probability,b,color="black",linewidth=0.85)
plt.grid()
plt.xlabel('y_hat', fontsize=14)
plt.ylabel('Binary Cross Entropy', fontsize=14)
plt.show()
```

```python
##Left side and right side of the inequality
##-> Don't forget to round the result :
##0.06000000005 != 0.06000000000
def leftSide (y_true,x,y,t):
  a = (1-t)*y+t*x
  return round(BinaryCrossEntropy(y_true,a),2)

def rightSide (y_true,x,y,t):
  a = (t*BinaryCrossEntropy(y_true,x))+((1-t)*BinaryCrossEntropy(y_true,y))
  return round(a,2)


#Convexity
y_expected_proba = 1
I = [0.01,1]
T  = np.arange(0,1,0.02)
X = np.arange(I[0],I[1],0.01)
Y = X[::-1]

nbTests = 0
nbPositivesTests = 0
for i in range(0,len(X)):
  x = X[i]
  y = Y[i]
  for t in T:
    nbTests+=1
    left = leftSide(y_expected_proba,x,y,t)
    right = rightSide(y_expected_proba,x,y,t)
    if (left < right or left==right):
      nbPositivesTests +=1
if (nbTests == nbPositivesTests):
  print("-> The Binary Cross Entropy function is convex")
else:
  print("-> The Binary Cross Entropy function is not convex")
```
