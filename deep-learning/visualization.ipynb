**4: Create an MSE Function**

And test your function with $\beta_0 = -44$ and $\beta_1 = 7.72$

```python
def MSE(y,predictions):
  n = len(y)
  sumErrors = 0
  for k in range(0, len(y)):
    sumErrors += (y[k]-predictions[k])**2
  return sumErrors/n
```

```python

beta_0 = -44
beta_1=7.72

predictions = []
for x in X:
  pred = x[0]*beta_1+beta_0
  predictions.append(pred)

print("MSE :",MSE(Y, predictions))
```

**4: Create the "Brut force" optimisation**

Calculate the MSE for all couple $\beta_1$,$\beta_0$ where $\beta_1[0,14]$ and $\beta_0[-86,0]$

```python
beta0_min = -86
beta0_max = 0

beta1_min = 0
beta1_max = 14

nbPoints = 900

betas0Values = np.linspace(start=beta0_min, stop=beta0_max, num=nbPoints)
betas1Values = np.linspace(start=beta1_min, stop=beta1_max, num=nbPoints)

graphBeta1, graphBeta0 = np.meshgrid(betas1Values, betas0Values)
graphMSE = np.zeros((nbPoints,nbPoints))

backup=[]

for i in range(nbPoints):
  for j in range (nbPoints):

    beta_1 = graphBeta1[i][j]
    beta_0 = graphBeta0[i][j]

    predictions = []
    for x in X:
      pred = x[0]*beta_1+beta_0
      predictions.append(pred)

    mse = MSE(Y,predictions)

    backup.append([beta_1,beta_0,mse])

    graphMSE[i][j] = mse
```

Find the minimum of MSE

```python
mseMIN = graphMSE.min()
coord = np.unravel_index(indices=graphMSE.argmin(), shape=graphMSE.shape)
i=coord[0]
j=coord[1]

Beta1MIN = graphBeta1[i][j]
Beta0MIN = graphBeta0[i][j]

print("MSE = ",mseMIN)
print("beta_0 = ",Beta0MIN)
print("beta_1 = ",Beta1MIN)
```
